#### Data 전처리 (코드 수정 X)

```python
data_numpy = np.array(data)

# 임시 입력 data 설정
x = np.zeros((data_numpy.shape[0], 3))
y = data_numpy[:, 0]
```

#### 거리 data 전처리

```python
# 좌표값을 사용하여 거리로 변환
delta_x = data_numpy[:, 3] - data_numpy[:, 1] # dropoff_x - pickup_x
delta_y = data_numpy[:, 4] - data_numpy[:, 2] # dropoff_y - pickup_y

# L1 distance 저장
x[:, 0] = np.abs(delta_x)+np.abs(delta_y) # 델타x + 델타y

# L2 distance 저장
x[:, 1] = np.sqrt(delta_x**2 + delta_y**2 )

# 승객 수 저장
x[:, 2] = data_numpy[:, -1]
```

#### Data min-max 정규화

```python
# x에 대한 min, max값 도출
x_min = np.min(x, axis=0)
x_max = np.max(x, axis=0)

# y에 대한 min, max값 도출
y_min = np.min(y)
y_max = np.max(y)

# 정규화 수행
x = (x-x_min) / (x_max-x_min)
y = (y-y_min) / (y_max-y_min)
```

#### Linear model 작성

```python
class LSM():
  def __init__(self):
    self.theta = None

  def fit(self, X, Y):
    N = X.shape[0] # N = 입력 데이터 개수

    # 입력 X에 대해 bias 차원 추가
    bias = np.ones((N, 1))
    X = np.hstack([X, bias])

    # Least Square Method 수행
    # X transpose 생성
    X_T = X.T
    # X^T * X 생성
    X_T_X = np.dot(X_T, X)
    # (X^T * X)^-1 생성
    X_T_X_inv = np.linalg.inv(X_T_X)
    # X^T * Y 생성
    X_T_Y = np.dot(X_T, Y)
    # theta 계산
    self.theta = np.dot(X_T_X_inv,X_T_Y)
    return self.theta

  def predict(self, X):
    N = X.shape[0] # N = 입력 데이터 개수

    # 입력 X에 대해 bias 차원 추가
    bias = np.ones((N, 1))
    X = np.hstack([X, bias])

    # 저장된 theta와 X를 이용하여 Y_hat 계산
    Y_hat = np.dot(X, self.theta)
    return Y_hat
```

#### weight, bias 학습 (코드 수정 X)

```python
linear_model = LSM()
theta = linear_model.fit(x, y)
print(theta)
```

#### 가격 예측

```python
l1_distance = 250
l2_distance = 197
passenger_count_test = 2

input_test = np.hstack([l1_distance, l2_distance, passenger_count_test])

# predict 함수가 2차원 배열을 요구하므로 (1, 3) 형태로 변환
input_test = input_test.reshape(1, 3)

# 신규 데이터 정규화 (데이터 전처리에서 사용한 min, max값 이용)
input_test = (input_test - x_min) / (x_max - x_min)

price_test = linear_model.predict(input_test)

# 출력 데이터 역정규화 (데이터 전처리에서 사용한 min, max값 이용)
price_test = price_test * (y_max - y_min) + y_min

print(price_test)
```
